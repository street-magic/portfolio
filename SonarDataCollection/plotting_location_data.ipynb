## Imports

```python
from rov import *
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
from matplotlib.axis import Axis  
```

## Data loading

```python
rov_location = LocationData("../00-Data/14.07.23/01-Raw/LOC-14.07.23-13.32.pcapng", nozeroes = False)
```

## Plotting location data

```python
x = np.array(rov_location.loc_times) - rov_location.loc_times[0]
t_start = 0
t_end = x[-1]
rov_x = np.array(rov_location.x) / 1000
rov_y = np.array(rov_location.y) / 1000
rov_z = np.array(rov_location.z) / 1000
fig, ax = plt.subplots(figsize=(12,12), nrows=3,ncols=2)
fps = len(rov_x)/x[-1]
start = math.floor(t_start * fps)
end = math.floor(t_end * fps)
ax[0][0].xaxis.set_major_formatter(FormatStrFormatter('%d s'))
ax[0][0].plot(x[start:end], rov_x[start:end], linewidth=.2, color = 'steelblue')
ax[0][0].set_title('X, m'),
ax[1][0].xaxis.set_major_formatter(FormatStrFormatter('%d s'))
ax[1][0].plot(x[start:end], rov_y[start:end], linewidth=.2, color = 'olive')
ax[1][0].set_title('Y, m'),
ax[2][0].plot(x[start:end], rov_z[start:end], linewidth=.2, color='brown')
ax[2][0].set_title('Z, m')
ax[2][0].xaxis.set_major_formatter(FormatStrFormatter('%d s'))
ax[0][1].plot(x[start:end], rov_location.rot_x[start:end], linewidth=.2, color = 'steelblue')
ax[0][1].set_title('Roll, deg')
ax[0][1].xaxis.set_major_formatter(FormatStrFormatter('%d s'))
ax[1][1].plot(x[start:end], rov_location.rot_y[start:end], linewidth=.2, color='olive')
ax[1][1].set_title('Pitch, deg')
ax[1][1].xaxis.set_major_formatter(FormatStrFormatter('%d s'))
ax[2][1].plot(x[start:end], rov_location.rot_z[start:end], linewidth=.2, color='brown')
ax[2][1].set_title('Yaw, deg')
ax[2][1].xaxis.set_major_formatter(FormatStrFormatter('%d s'))
plt.show()
```
