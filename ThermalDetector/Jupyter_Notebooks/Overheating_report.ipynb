```python
import matplotlib.pyplot as plt
from matplotlib import cm as _pltcolormap
from matplotlib import colors as _pltcolors
import numpy as np
import pandas as pd
import sys
import pathlib
import os
```

```python
colormap = _pltcolors.LinearSegmentedColormap.from_list(
    "", ['#0008e3', '#6aed37', '#e3000f'])
norm = _pltcolors.Normalize(
    vmin=22, vmax=33.75, clip=True)
```

```python
raw_old_df = pd.read_csv(r'C:\Users\WW\Desktop\2022-09-03_15-17-13\EYE_old.csv')
raw_new_df = pd.read_csv(r'C:\Users\WW\Desktop\2022-09-03_15-17-13\EYE_new.csv')

old = raw_old_df.drop(['received_time'], axis=1)
new = raw_new_df.drop(['received_time'], axis=1)
```

```python
print (max(old.max().values))
print (min(old.min().values))

print (max(new.max().values))
print (min(new.min().values))
```

```python
col_num = 8
row_num = old.shape[0] // col_num + (1 if old.shape[0] % col_num >= 1 else 0)
T = 0.75

fig, axs = plt.subplots(row_num, col_num, figsize=(col_num * 2, row_num * 2))
plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[])
n = old.shape[0]
for i in range(n):
    x = i // col_num
    y = i % col_num
    #diff = (raw_df_nolabels.iloc[i] - bg_mean).values.reshape(8,8)
    #bg_removed = np.where(diff > T, diff, 0)
    axs[x,y].imshow(old.iloc[i].values.reshape(8,8), cmap=colormap, norm=norm)
```
